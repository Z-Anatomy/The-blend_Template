```python
import pandas as pd

df = pd.read_csv('./TA2.csv')

df
```

```python
LIST_3D = []
LIST_DEFINITIONS = []
LIST_LABELS = []

lists = ['./List_3D', './List_Definitions', './List_labels']


for i,filename in enumerate(lists):
    with open(filename) as f:
        lines = f.readlines()
    for line in lines: 
        line =  line.replace("\n", "")
        if i == 0:
            LIST_3D.append(line)
        if i == 1:
            LIST_DEFINITIONS.append(line)
        if i == 2:
            LIST_LABELS.append(line)



```

```python
d3 = []
lable = []
Def = []


for row in df['English']:
    progress = ''
    if row in LIST_3D:
        progress = " x"
        d3.append(progress)
    else:
        d3.append('')        

    if row in LIST_LABELS:
        progress = " x"
        lable.append(progress)
    else:
        lable.append('') 

    if row in LIST_DEFINITIONS:
        progress = " x"
        Def.append(progress)
    else:
        Def.append('') 
    
        
print(len(d3))

df['3D'] = d3
df['label'] = lable
df['Def'] = Def

print(df)

df.to_csv('./TA2Progress.csv')





  
```
